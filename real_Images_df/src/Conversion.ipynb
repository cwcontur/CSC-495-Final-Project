```python
import pandas as pd
import numpy as np

df=pd.read_csv(r"C:\Users\saive\Documents\github\CSC-495-Final-Project\real_Images_df\data\original.csv")
df
```

```python
df_group=df.groupby(['Name'])
df_group.first()
```

```python
df_final=pd.DataFrame()
for group in df_group:
    group_name=group[0]
    df_new=pd.DataFrame()
    df_new['Name']=[group_name]
    df_1=df_group.get_group(group_name).reset_index()
    df_1=df_1.transpose()
    #print(df_1)
    for i in range(len(df_1.loc['label'])):
        s=df_1.loc['label'][i]
        s1=s+'_x'
        s2=s+'_y'
        df_new[s1]=np.array([df_1.loc['x'][i]])
        df_new[s2]=np.array([df_1.loc['y'][i]])
    df_final=pd.concat([df_final,df_new],axis=0)
df_final
```
