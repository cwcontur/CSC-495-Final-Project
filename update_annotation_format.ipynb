```python
import os
from PIL import Image
import cv2
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import matplotlib.patches as patches
from tqdm.notebook import tqdm
```

```python
BASE_DIR = os.path.dirname(os.path.abspath("__file__"))
DATA_DIR = os.path.join(BASE_DIR, "data")
MPII_annotation = os.path.join(DATA_DIR, "mpii_dataset.csv")
```

```python
pd_data = pd.read_csv(MPII_annotation)
pd_data.drop(pd_data.columns[0], axis=1, inplace=True)
```

```python
pd_data
```

```python
pd_data = pd.DataFrame(pd_data.apply(lambda x: {
    "image_name": x['NAME'],
    "left_shoulder": (x['l shoulder_X'], x['l shoulder_Y']),
    "right_shoulder": (x['r shoulder_X'], x['r shoulder_Y']),
    "left_elbow": (x['l elbow_X'], x['l elbow_Y']),
    "right_elbow": (x['r elbow_X'], x['r elbow_Y']),
    "left_wrist": (x['l wrist_X'], x['l wrist_Y']),
    "right_wrist": (x['r wrist_X'], x['r wrist_Y']),
    "left_hip": (x['l hip_X'], x['l hip_Y']),
    "right_hip": (x['r hip_X'], x['r hip_Y']),
    "left_knee": (x['l knee_X'], x['l knee_Y']),
    "right_knee": (x['r knee_X'], x['r knee_Y']),
    "left_ankle": (x['l ankle_X'], x['l ankle_Y']),
    "right_ankle": (x['r ankle_X'], x['r ankle_Y']),
    "pelvis": (x['pelvis_X'], x['pelvis_Y']),
    "thorax": (x['thorax_X'], x['thorax_Y']),
    "upper_neck": (x['upper neck_X'], x['upper neck_Y']),
    "head_top": (x['head top_X'], x['head top_Y']),
    "scale": x['Scale'],
    "activity": x['Activity'],
    "category": x['Category']}, axis=1).to_list())

updated = "/".join(MPII_annotation.split("/")[:-1]) + "/updated_" + MPII_annotation.split("/")[-1]
pd_data.to_csv(updated, index=False)
```
